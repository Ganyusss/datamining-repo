```python
import folium
from folium.plugins import HeatMap
```

```python
terror = pd.read_csv('jup\\Chicago Crimes.csv')
```

<h1>Insight 1</h1>
<h4>Arson</h4>

```python
arson = terror[(terror['Primary Type'] == "DECEPTIVE PRACTICE") & (terror['Arrest'] == True)]
arsondf = arson.groupby(['Longitude','Latitude']).size().reset_index(name = 'incident_count')
arsondf

arsondf['normalized_weight'] = (arsondf['incident_count'] - arsondf['incident_count'].min()) / \
                                     (arsondf['incident_count'].max() - arsondf['incident_count'].min())

arsondf = arsondf[[ 'Latitude', 'Longitude','normalized_weight']].values.tolist()

ars = folium.Map(location=[ 41.8781, -87.6298], zoom_start=10)

HeatMap(arsondf).add_to(ars)

ars.save('arson.html')
ars
```
