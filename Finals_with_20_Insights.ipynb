<h1> Analysis

<h1> 1. How does Total Household Income vary across different Regions?

```python
plt.figure(figsize=(20,6))
sns.violinplot(x='Region', y='Total Household Income', data=family_finance)
plt.title('Distribution of Total Household Income by Region')
plt.xticks(rotation=45)
plt.show()
```

<h1> Insight 1 </h1>

<h2> The distribution of Total Household Income highlights clear regional disparities across the Philippines. While the majority of households in most regions experience low to moderate income levels, Regions VI (Western Visayas) and VII (Eastern Visayas) exhibit notable high-income outliers. The National Capital Region (NCR) demonstrates the greatest income diversity with a wide range of incomes, whereas the Autonomous Region in Muslim Mindanao (ARMM) shows the most compressed income distribution. These differences likely stem from varying levels of economic development, urbanization, and industry presence across regions. </h2>

<h1> 2. Does the Main Source of Income influence Total Household Income within regions?

```python
plt.figure(figsize=(15,6))
sns.barplot(x='Main Source of Income', y='Total Household Income', hue='Region', data=family_finance)
plt.title('Average Total Household Income by Main Source of Income across Regions')
plt.xticks(rotation=45)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

<h1> Insight 2 </h1>

<h2> It does influences the total household income across regions. Households relying on entrepreneurial activities and other sources of income tend to have higher average incomes compared to those dependent solely on wages or salaries. Additionally, there is notable regional variation within each income source category, with regions like NCR consistently showing higher income levels across sources. This suggests that both the nature of income and regional economic conditions play important roles in determining household income. </h2>

<h1> 3.  Given income levels, does Total Food Expenditure relate to Total Household Income?

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Total Household Income', y='Total Food Expenditure', data=family_finance, alpha=0.6)
plt.title('Total Food Expenditure vs. Total Household Income')
plt.show()
```

<h1> Insight 3 </h1>

<h2> Total Food Expenditure has a positive correlation with Total Household Income at lower income levels, where food spending rises with income. However, this link weakens among higher-income households, who show varied food spending regardless of income increases, suggesting a saturation point in food expenditure for wealthier families. </h2>

<h1> 4. Which food category contributes most to Total Food Expenditure across income levels?

```python
food_categories = ['Bread and Cereals Expenditure', 'Total Rice Expenditure', 'Meat Expenditure',
                   'Total Fish and  marine products Expenditure', 'Fruit Expenditure', 'Vegetables Expenditure']

food_means = family_finance[food_categories].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x=food_means.index, y=food_means.values)
plt.title('Average Expenditure by Food Categories')
plt.xticks(rotation=45)
plt.show()
```

<h1> Insight 4 </h1>

<h2> Bread and Cereals account for the largest share of household food spending, followed by Rice, underscoring the importance of staple grains in the diet. Meat and Fish rank next with similar spending levels, while Vegetables and especially Fruits receive the smallest allocations. This pattern highlights a dietary focus on carbohydrate-rich staples, with moderate spending on protein and minimal on fruits. </h2>

<h1> 5. How does the Agricultural Household indicator affect the expenditure pattern on food?

```python
family_finance['Agricultural Household indicator'].value_counts()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='Agricultural Household indicator', y='Total Food Expenditure', data=family_finance)
plt.title('Total Food Expenditure by Household Type')
plt.xticks([0, 1, 2], ['Non-Agricultural', 'Agricultural', 'Mixed'])
plt.show()
```

<h1> Insight 5 </h1>

<h2> Food expenditure varies notably by household type. Mixed households have the highest median spending and widest range, followed by Non-Agricultural households. Agricultural households spend the least and show a narrower distribution. High outliers appear in all groups, but are most frequent among Non-Agricultural households. These differences may reflect income disparities, access to food, and the degree of reliance on homegrown versus purchased goods. </h2>

<h1> 9. How does the Type of Household influence the Total Number of Family Members and their employment?

```python
plt.figure(figsize=(14,6))
sns.boxplot(x='Type of Household', y='Total Number of Family members', data=family_finance)
plt.title('Family Size by Type of Household')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(14,6))
sns.boxplot(x='Type of Household', y='Total number of family members employed', data=family_finance)
plt.title('Number of Employed Family Members by Type of Household')
plt.xticks(rotation=45)
plt.show()
```

<h1> Insight 9 </h1>

<h2> The charts show how family size and employment vary by household type. Extended families and those with two or more household representatives tend to have larger family sizes compared to single-family households. Notably, households with multiple representatives also report the highest number of employed family members, suggesting that larger or more complex households may have greater earning potential. In contrast, single-family households typically have fewer members and a lower count of employed individuals, potentially reflecting more limited workforce resources.  </h2>

<h1> 10. What is the relationship between the number of employed family members and total household income?

```python
plt.figure(figsize=(10,6))
sns.barplot(
    x='Total number of family members employed',
    y='Total Household Income',
    data=family_finance,
    estimator=np.mean,
    ci=None
)
plt.title('Average Household Income by Number of Employed Family Members')
plt.xlabel('Number of Employed Family Members')
plt.ylabel('Average Total Household Income')
plt.show()
```

<h1> Insight 10 </h1>

<h2> Average household income increases with the number of employed family members, particularly between two and five earners. The trend flattens beyond five, suggesting possible limits due to job quality or time constraints. Notably, even households with no employed members report income likely from remittances, pensions, or other non-labor sources emphasizing the role of diversified income streams in household finances. </h2>
