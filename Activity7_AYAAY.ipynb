<h2> Importing libraries

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import folium
from folium.plugins import HeatMap

import warnings
warnings.filterwarnings('ignore')
```

<h2> Import dataframe

```python
terror = pd.read_csv('jup\\Global_terrorist.csv', encoding='ISO-8859-1')
```

```python
terror
```

<h1>INSIGHTS</h1>

<h1>Example provided</h1>

```python


filtered_data = terror[(terror['region_txt'] == 'Middle East & North Africa') & 
                     (terror['iyear'] >= 2000) & 
                     (terror['attacktype1_txt'] == 'Bombing/Explosion') & 
                     (terror['nkill'] > 10)]

aggregated_data = filtered_data.groupby(['latitude', 'longitude']).size().reset_index(name='incident_count')

aggregated_data['normalized_weight'] = (aggregated_data['incident_count'] - aggregated_data['incident_count'].min()) / \
                                       (aggregated_data['incident_count'].max() - aggregated_data['incident_count'].min())

heat_data = aggregated_data[['latitude', 'longitude', 'normalized_weight']].values.tolist()

m = folium.Map(location=[30, 40], zoom_start=4)

HeatMap(heat_data).add_to(m)

m.save('terrorism_heatmap.html')
m
```

<h1> Insight 1 Regional Analysis</h1>


<h2>"Where have assassination incidents been most concentrated in South America since 1980?"</h2>


<h3> This analysis generates a heatmap that visualizes the geographical distribution of assassination-related terrorist incidents in South America from 1980 onward. </h3>

```python
filtered_df = terror[(terror['region_txt'] == 'South America') &(terror['iyear'] >= 1980) &(terror['attacktype1_txt'] == 'Assassination')]

aggregated_df = filtered_df.groupby(['latitude', 'longitude']).size().reset_index(name='incident_count')
aggregated_df['normalized_weight'] = ((aggregated_df['incident_count'] - aggregated_df['incident_count'].min()) /(aggregated_df['incident_count'].max() - aggregated_df['incident_count'].min()))

heat_data = aggregated_df[['latitude', 'longitude', 'normalized_weight']].values.tolist()
assasinate = folium.Map(location=[-15, -60], zoom_start=3)
HeatMap(heat_data).add_to(assasinate)
assasinate.save('assasinate.html')
assasinate
```
